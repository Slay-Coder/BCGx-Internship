# Box Plot for Price Sensitivity Analysis

Energy price sensitivity in relation with Churning status

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example DataFrame creation (replace with your data)
# mergedData = pd.read_csv('path_to_merged_data.csv')

# Replace 0 and 1 in 'churn' with descriptive labels
mergedData['churning'] = mergedData['churn'].map({0: 'No Churn', 1: 'Churn'})

# Select relevant columns and melt the DataFrame
price_cols = ['price_off_peak_var', 'price_mid_peak_var', 'price_peak_var']
melted_data = pd.melt(mergedData, id_vars=['churn'], value_vars=price_cols,
                      var_name='price_period', value_name='price')

# Rename the price periods for better readability
price_period_mapping = {
    'price_off_peak_var': 'Off-Peak',
    'price_mid_peak_var': 'Mid-Peak',
    'price_peak_var': 'Peak'
}
melted_data['price_period'] = melted_data['price_period'].map(price_period_mapping)

# Box Plot for Price Sensitivity Analysis with hue for different periods
plt.figure(figsize=(14, 8))
sns.boxplot(x='churn', y='price', hue='price_period', data=melted_data, color="green",dodge=True)
plt.title('Distribution of Variable Prices by Churn Status and Period')
plt.xlabel('Churning Status')
plt.ylabel('Variable Price')
plt.legend(title='Price Period')
plt.show()
```
